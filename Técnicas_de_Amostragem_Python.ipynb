T√©cnicas de Amostragem, temos 3 principais :

- **T√©cnica de amostragem aleat√≥ria simples** :
    - T√©cnica de amostragem em que todos os indiv√≠duos de uma popula√ß√£o t√™m a mesma chance de serem selecionados para a amostra.
- **Amostragem Sistem√°tica :**
    - √â como a amostragem aleat√≥ria, contudo h√° um fator matem√°tico de sele√ß√£o que a diferencia.
    - Os elementos s√£o divididos em grupos numericamente iguais, ent√£o √© definido um ‚Äúponto de partida‚Äù, com intuito de estabelecer um n√∫mero que se repetir√°
        
        <aside>
        üí° Por exemplo, em uma pesquisa realizada em um bairro, podemos dividir <b>500 moradias</b> em <b>50 grupos</b>, para obter uma amostragem com 50 resid√™ncias ,dessa maneira, <b>a cada 10 domic√≠lios 1 pesquisa</b> ser√° realizada.
        
        Dentro desse grupo um ‚Äú<b>ponto de partida</b>‚Äù, ou seja, um n√∫mero, deve ser selecionado. Assim, se decidirmos que o **4¬∞ domic√≠lio** de um grupo ser√° o selecionado, vamos obter uma sequ√™ncia(**4, 14, 24, 34‚Ä¶ at√© 494**) determinada de resid√™ncias para a amostragem.
        
        </aside>

    - T√©cnica realizada em duas etapas :
        1. Dividir a popula√ß√£o em grupos distintos
